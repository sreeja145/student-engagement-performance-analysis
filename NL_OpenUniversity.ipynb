## Data Merging

```python
# 1)Base table: one row per student–module–presentation
# ---------------------------------------------------
base = student_info.copy()

# Adding registration info
base = base.merge(
    student_reg[["code_module", "code_presentation",
                         "id_student", "date_registration", "date_unregistration"]],
    on=["code_module", "code_presentation", "id_student"],
    how="left"
)

# ---------------------------------------------------
# 2) Assessments: attach module/presentation, then aggregating
# ---------------------------------------------------

# 2.1 Attaching code_module & code_presentation to studentAssessment via assessments
sa_with_mod = student_assess.merge(
    assessments[["id_assessment", "code_module", "code_presentation"]],
    on="id_assessment",
    how="left"
)

# 2.2 Aggregating assessment info per student–module–presentation
sa_agg = (
    sa_with_mod
    .groupby(["code_module", "code_presentation", "id_student"])
    .agg(
        n_assessments=("id_assessment", "nunique"),
        mean_score=("score", "mean")
    )
    .reset_index()
)

# 2.3 Merge assessment aggregates into base
base = base.merge(
    sa_agg,
    on=["code_module", "code_presentation", "id_student"],
    how="left"
)

# ---------------------------------------------------
# 3) VLE: optionally enrich with activity_type, then aggregate
# ---------------------------------------------------

# 3.1 Attach activity_type to studentVle via vle
sv_with_act = student_vle.merge(
    vle[["id_site", "code_module", "code_presentation", "activity_type"]],
    on=["id_site", "code_module", "code_presentation"],
    how="left"
)

# 3.2 Aggregate VLE features per student–module–presentation
sv_agg = (
    sv_with_act
    .groupby(["code_module", "code_presentation", "id_student"])
    .agg(
        total_clicks=("sum_click", "sum"),
        active_days=("date", "nunique")
    )
    .reset_index()
)

# 3.3 Merge VLE aggregates into base
base = base.merge(
    sv_agg,
    on=["code_module", "code_presentation", "id_student"],
    how="left"
)

# ---------------------------------------------------
# 4) Result: modelling table
# ---------------------------------------------------

model_df = base.copy()
# model_df now has demographics + registration + assessment agg + VLE agg,
# with final_result from studentInfo as the target.

print(model_df.shape)
print(model_df.columns)
```

# Academic Performance Metric

```python
import numpy as np

# ============================
# 1) Preparing assessment-level data
# ============================

# Attaching module, presentation, assessment_type, weight
sa = student_assess.merge(
    assessments[["id_assessment",
                 "code_module",
                 "code_presentation",
                 "assessment_type",
                 "weight"]],
    on="id_assessment",
    how="left"
)

# Pass/Fail per assessment based on score threshold
PASS_THRESHOLD = 40
sa["passed_assessment"] = (sa["score"] >= PASS_THRESHOLD).astype(int)

# Weighted score per assessment (score * weight)
sa["weighted_score"] = sa["score"].fillna(0) * sa["weight"].fillna(0)

# ============================
# Aggregate to student–module–presentation level
# ============================

sa_ap = (
    sa.groupby(["code_module", "code_presentation", "id_student"])
      .agg(
          # from studentAssessment.score
          ap_mean_score=("score", "mean"),           # average score over all assessments
          ap_n_assigned=("id_assessment", "nunique"),
          ap_n_submitted=("score", "count"),

          # number of passed assessments per module
          ap_n_passed=("passed_assessment", "sum"),

          # weighted score across assessment types (TMA, CMA, Exam)
          ap_weighted_score=("weighted_score", "sum"),
      )
      .reset_index()
)

# ============================
#  Merging AP ingredients into  merged student table
# ============================

df = model_df.merge(
    sa_ap,
    on=["code_module", "code_presentation", "id_student"],
    how="left"
)

# ============================
# Constructing Academic Performance (AP)
#    using only allowed components:
#    - studentAssessment.score
#    - Pass/Fail per assessment
#    - final_result (Pass, Distinction, Fail, Withdrawn)
#    - number of passed assessments per module
#    - weighted score across assessment types
# ============================

#  Normalized Average Score per student (0–1)
df["ap_norm_score"] = (df["ap_mean_score"] / 100.0).fillna(0)

#  Completion indicator (1 = course has any final outcome; 0 = no final outcome)
#     Includes Pass, Distinction, Fail, Withdrawn as completed outcomes.
df["ap_completed"] = df["final_result"].isin(
    ["Pass", "Distinction", "Fail", "Withdrawn"]
).astype(int)

#  Number of passed assessments per module (normalized)
max_pass = df["ap_n_passed"].max()
if pd.notna(max_pass) and max_pass > 0:
    df["ap_pass_count_norm"] = (df["ap_n_passed"] / max_pass).fillna(0)
else:
    df["ap_pass_count_norm"] = 0

#  Weighted score across assessment types (normalized)
max_w = df["ap_weighted_score"].max()
if pd.notna(max_w) and max_w > 0:
    df["ap_norm_weighted"] = (df["ap_weighted_score"] / max_w).fillna(0)
else:
    df["ap_norm_weighted"] = 0

#  Final AP_raw (we can adjust these weights, but only these components are used)
df["AP_raw"] = (
    0.4 * df["ap_norm_score"] +
    0.3 * df["ap_completed"] +
    0.2 * df["ap_pass_count_norm"] +
    0.1 * df["ap_norm_weighted"]
)

# df now contains a properly defined Academic Performance measure in column "AP_raw"
```

# Learning Effort Metric

```python
import numpy as np

# ========= 1) Total VLE clicks & Active VLE days (already aggregated) =========
# We already have:
#   - merged_df["total_clicks"]
#   - merged_df["active_days"]
# We rename them into LE components:

le_df = model_df.copy()
le_df["le_total_clicks"] = le_df["total_clicks"].fillna(0)
le_df["le_active_days"]  = le_df["active_days"].fillna(0)

# ========= 2) Assessment submission count (attempts per assessment) =========
# Each row in student_assess is one submission; counting rows gives submission count.

sa_sub = (
    student_assess
    .merge(
        assessments[["id_assessment", "code_module", "code_presentation"]],
        on="id_assessment",
        how="left"
    )
    .groupby(["code_module", "code_presentation", "id_student"])
    .size()
    .reset_index(name="le_submissions")   # total submissions for that student-module
)

le_df = le_df.merge(
    sa_sub,
    on=["code_module", "code_presentation", "id_student"],
    how="left"
)
le_df["le_submissions"] = le_df["le_submissions"].fillna(0)

# ========= 3) Registration duration (time window of study) =========
# Using studentRegistration dates already merged into merged_df:
# date_unregistration can be NaN -> use course-end proxy (module median duration).

le_df["reg_duration_raw"] = le_df["date_unregistration"] - le_df["date_registration"]

# Module-presentation median duration as fallback
mod_reg_med = le_df.groupby(
    ["code_module", "code_presentation"]
)["reg_duration_raw"].transform(
    lambda x: x.replace(0, np.nan).median()
)

le_df["le_reg_duration"] = le_df["reg_duration_raw"].fillna(mod_reg_med)

# for NaN (all missing in group), fill with global median
global_reg_med = le_df["le_reg_duration"].median()
le_df["le_reg_duration"] = le_df["le_reg_duration"].fillna(global_reg_med)

# ========= 4) Number of enrolled modules per student =========
# Based only on student_info (how many module-presentations a student is in).

mods_per_student = (
    student_info.groupby("id_student")[["code_module", "code_presentation"]]
    .nunique()
    .sum(axis=1)
    .rename("le_n_modules")
    .reset_index()
)

le_df = le_df.merge(mods_per_student, on="id_student", how="left")
le_df["le_n_modules"] = le_df["le_n_modules"].fillna(1)

# ========= 5) Normalize each LE component =========

# (a) Normalized total clicks per module-presentation
mod_click_med = le_df.groupby(
    ["code_module", "code_presentation"]
)["le_total_clicks"].transform("median")

le_df["le_norm_clicks"] = (
    le_df["le_total_clicks"] / mod_click_med.replace(0, np.nan)
).fillna(0)

# (b) Active Study Days
mod_days_med = le_df.groupby(
    ["code_module", "code_presentation"]
)["le_active_days"].transform("median")

le_df["le_norm_days"] = (
    le_df["le_active_days"] / mod_days_med.replace(0, np.nan)
).fillna(0)

# (c) Normalized registration duration (time from registration to end/withdrawal)
mod_reg_dur_med = le_df.groupby(
    ["code_module", "code_presentation"]
)["le_reg_duration"].transform("median")

le_df["le_norm_regdur"] = (
    le_df["le_reg_duration"] / mod_reg_dur_med.replace(0, np.nan)
).fillna(1)

# (d) Normalized assessment submission count (attempts per assessment)
max_sub = le_df["le_submissions"].max()
le_df["le_norm_submissions"] = (
    le_df["le_submissions"] / max_sub
).fillna(0) if max_sub and max_sub > 0 else 0

# (e) Normalized number of enrolled modules
max_mods = le_df["le_n_modules"].max()
le_df["le_norm_n_modules"] = (
    le_df["le_n_modules"] / max_mods
).fillna(0) if max_mods and max_mods > 0 else 0

# ========= 6) Final LE_raw  =========

le_df["LE_raw"] = (
    0.4 * le_df["le_norm_clicks"] +      # Normalized Total Clicks
    0.2 * le_df["le_norm_days"] +        # Active Study Days
    0.2 * le_df["le_norm_regdur"] +      # Time from registration to end
    0.1 * le_df["le_norm_submissions"] + # Assessment submission count
    0.1 * le_df["le_norm_n_modules"]     # Number of enrolled modules
)

# le_df now has a Learning Effort measure "LE_raw"
# from: total clicks, active days, submissions, registration duration, num modules.
```

# Academic Success Efficiency (ASE)
Academic Success Efficiency (ASE) = Academic Performance ÷ Learning Effort



```python
# handling missing or negative values:
ase_df["AP_raw"] = ase_df["AP_raw"].fillna(0).clip(lower=0)
ase_df["LE_raw"] = ase_df["LE_raw"].fillna(0).clip(lower=0)

# keeping AP and LE on a common scale
# Raw AP and LE are on different scales (e.g., AP might be around 0–1, LE around 0–2).
#  normalize within each module‑presentation, so “1.0” means “average for this module”.
# min–max scaling to 0–1 inside each module:

group_keys = ["code_module", "code_presentation"]

def to_01(s):
    s_min = s.min()
    s_max = s.max()
    if s_max == s_min:          # everyone has same value
        return pd.Series(0.5, index=s.index)
    return (s - s_min) / (s_max - s_min)

ase_df["AP_norm"] = (
    ase_df.groupby(group_keys)["AP_raw"].transform(to_01)
)
ase_df["LE_norm"] = (
    ase_df.groupby(group_keys)["LE_raw"].transform(to_01)
)
```

### Interpretation:
### AP_norm = 0 → lowest performance in that module
### AP_norm = 1 → highest performance in that module
#### LE_norm similar

```python
# Computing ASE as a ratio
# Now ASE is just normalized performance divided by normalized effort:
eps = 1e-3 # to avoid division by zero (we are adding 0.001)
ase_df["ASE_score"] = ase_df["AP_norm"] / (ase_df["LE_norm"] + eps)
ase_df["ASE_score"] = ase_df["ASE_score"].clip(lower=0, upper=5)
```

ASE_score:

Around 1 → performance and effort are in balance (average AP for average LE).

> 1 → higher performance than you’d expect for that level of effort (efficient).

< 1 → lower performance for that effort (inefficient / at risk).

```python

bins = [-float("inf"), 0.75, 1.25, float("inf")]
labels = ["Low ASE", "Balanced ASE", "High ASE"]
ase_df["ASE_band"] = pd.cut(ase_df["ASE_score"], bins=bins, labels=labels)
```

## Assessment Effectiveness Metrics

```python

# Attaching module and assessment_type to each student attempt
sa = student_assess.merge(
    assessments[["id_assessment","code_module","code_presentation","assessment_type","date"]],
    on="id_assessment",
    how="left"
)

# Pass/fail per attempt (same threshold as AP)
PASS_THRESHOLD = 40
sa["passed"] = (sa["score"] >= PASS_THRESHOLD).astype(int)

# 1) Pass Rate = Passed / Total assessments (per module-presentation)
assess_eff_mod = (
    sa.groupby(["code_module","code_presentation"])
    .agg(
        total_assessments=("id_assessment","count"),
        passed_assessments=("passed","sum"),
        pass_rate=("passed","mean")
    )
    .reset_index()
)

# 2) Average Score by Assessment Type (TMA, CMA, Exam) per module-presentation
avg_score_type = (
    sa.groupby(["code_module","code_presentation","assessment_type"])["score"]
    .mean()
    .reset_index()
    .rename(columns={"score":"avg_score_by_type"})
)

# 3) Submission Rate = Submitted / Assigned (per module-presentation)
# "Assigned" = distinct idassessment in that module-presentation
assigned = (
    assessments.groupby(["code_module","code_presentation"])["id_assessment"]
    .nunique()
    .reset_index()
    .rename(columns={"id_assessment":"n_assigned"})
)
submitted = (
    sa.groupby(["code_module","code_presentation"])["id_assessment"]
    .nunique()
    .reset_index()
    .rename(columns={"id_assessment":"n_submitted"})
)
sub_rate = assigned.merge(submitted, on=["code_module","code_presentation"], how="left")
sub_rate["n_submitted"] = sub_rate["n_submitted"].fillna(0)
sub_rate["submission_rate"] = sub_rate["n_submitted"] / sub_rate["n_assigned"]

# 4) Performance Trend = mean score by attempt order
sa_sorted = sa.sort_values(["code_module","code_presentation","id_student","date_submitted"])
sa_sorted["attempt_num"] = sa_sorted.groupby(
    ["code_module","code_presentation","id_student"]
).cumcount() + 1

perf_trend = (
    sa_sorted.groupby(["code_module","code_presentation","attempt_num"])["score"]
    .mean()
    .reset_index()
    .rename(columns={"score":"avg_score_by_attempt"})
)
```

```python


# -------------------------------------------------------------------
# 2) Early Engagement Score = Clicks within first 30 days
#    Using studentvle (date) aggregated per student–module
# -------------------------------------------------------------------
early_window = 30

early_clicks = (
    student_vle[student_vle["date"] <= early_window]
    .groupby(["code_module","code_presentation","id_student"])["sum_click"]
    .sum()
    .reset_index()
    .rename(columns={"sum_click":"early_clicks"})
)

eng_df = eng_df.merge(early_clicks,
    on=["code_module","code_presentation","id_student"],
    how="left"
)
eng_df["early_clicks"] = eng_df["early_clicks"].fillna(0)
```

```python


# -------------------------------------------------------------------
# 3) Content Coverage Rate =
#    Unique VLE activities accessed ÷ Total available activities
# -------------------------------------------------------------------
# 3.1 Total available activities per module-presentation (from vle)
total_activities = (
    vle.groupby(["code_module","code_presentation"])["id_site"]
    .nunique()
    .reset_index()
    .rename(columns={"id_site":"n_activities_available"})
)

# 3.2 Unique activities each student accessed (from studentvle)
student_activities = (
    student_vle.groupby(["code_module","code_presentation","id_student"])["id_site"]
    .nunique()
    .reset_index()
    .rename(columns={"id_site":"n_activities_accessed"})
)

eng_df = eng_df.merge(total_activities,
                      on=["code_module","code_presentation"],
                      how="left")

eng_df = eng_df.merge(student_activities,
                      on=["code_module","code_presentation","id_student"],
                      how="left")

eng_df["n_activities_accessed"] = eng_df["n_activities_accessed"].fillna(0)
eng_df["content_coverage_rate"] = (
    eng_df["n_activities_accessed"] /
    eng_df["n_activities_available"].replace(0, np.nan)
).fillna(0)
```

```python
# -------------------------------------------------------------------
# 4) Assessment Preparation Ratio =
#    Clicks before assessment date ÷ Total clicks
#    (approximation: clicks on days <= min assessment date for module)
# -------------------------------------------------------------------
# 4.1 Earliest assessment date per module-presentation
first_assess_date = (
    assessments.groupby(["code_module","code_presentation"])["date"]
    .min()
    .reset_index()
    .rename(columns={"date":"first_assess_day"})
)

# 4.2 Clicks before that day per student-module
sa_prep = student_vle.merge(
    first_assess_date,
    on=["code_module","code_presentation"],
    how="left"
)

prep_clicks = (
    sa_prep[sa_prep["date"] <= sa_prep["first_assess_day"]]
    .groupby(["code_module","code_presentation","id_student"])["sum_click"]
    .sum()
    .reset_index()
    .rename(columns={"sum_click":"prep_clicks"})
)

eng_df = eng_df.merge(
    prep_clicks,
    on=["code_module","code_presentation","id_student"],
    how="left"
)
eng_df["prep_clicks"] = eng_df["prep_clicks"].fillna(0)

eng_df["prep_ratio"] = (
    eng_df["prep_clicks"] /
    eng_df["total_clicks"].replace(0, np.nan)
).fillna(0)

# eng_df now has Engagement Health metrics:
# - vle_intensity
# - early_clicks
# - content_coverage_rate
# - prep_ratio
```

## Retention Risk Indicators

#### model_with_ASE should have the columns we need: AP_raw, LE_raw, early_clicks, totalclicks, registration dates, and so on

```python
# 1. Start from modeldf
model_with_ASE = model_df.copy()

# 2. Add AP_raw and AP components
model_with_ASE = model_with_ASE.merge(
    df[["code_module","code_presentation","id_student",
        "AP_raw","ap_n_assigned","ap_n_submitted","ap_n_passed"]],
    on=["code_module","code_presentation","id_student"],
    how="left"
)

# 3. Add LE_raw
model_with_ASE = model_with_ASE.merge(
    le_df[["code_module","code_presentation","id_student","LE_raw"]],
    on=["code_module","code_presentation","id_student"],
    how="left"
)

# 4. Add Engagement Health metrics from eng_df
model_with_ASE = model_with_ASE.merge(
    eng_df[["code_module","code_presentation","id_student",
            "vle_intensity","early_clicks",
            "n_activities_available","n_activities_accessed",
            "content_coverage_rate","prep_clicks","prep_ratio"]],
    on=["code_module","code_presentation","id_student"],
    how="left"
)

# 5. Add ASE_score
model_with_ASE = model_with_ASE.merge(
    ase_df[["code_module","code_presentation","id_student","ASE_score"]],
    on=["code_module","code_presentation","id_student"],
    how="left"
)
```

```python
# Use the AP counts we merged from df
model_with_ASE["submission_rate_student"] = (
    model_with_ASE["ap_n_submitted"] /
    model_with_ASE["ap_n_assigned"].replace(0, pd.NA)
)
model_with_ASE["submission_rate_student"] = (
    model_with_ASE["submission_rate_student"].fillna(0)
)
```

```python
# Target: 1 if Withdrawn, else 0
model_with_ASE["is_withdrawn"] = (
    model_with_ASE["final_result"] == "Withdrawn"
).astype(int)

# ============================================================
# 1) Low Early Engagement + Low Submission Rate
# ============================================================

early_q1 = model_with_ASE["early_clicks"].quantile(0.25)
sub_q1   = model_with_ASE["submission_rate_student"].quantile(0.25)

model_with_ASE["low_early_engagement"] = (
    model_with_ASE["early_clicks"] <= early_q1
).astype(int)

model_with_ASE["low_submission_rate"] = (
    model_with_ASE["submission_rate_student"] <= sub_q1
).astype(int)

model_with_ASE["risk_lowEarly_lowSub"] = (
    (model_with_ASE["low_early_engagement"] == 1) &
    (model_with_ASE["low_submission_rate"] == 1)
).astype(int)

# ============================================================
# 2) High Effort but Declining Scores
#    - high effort = top 25% of LE_raw
#    - low performance = bottom 25% of AP_raw
# ============================================================

le_q3 = model_with_ASE["LE_raw"].quantile(0.75)
ap_q1 = model_with_ASE["AP_raw"].quantile(0.25)

model_with_ASE["high_effort"] = (
    model_with_ASE["LE_raw"] >= le_q3
).astype(int)

model_with_ASE["low_AP"] = (
    model_with_ASE["AP_raw"] <= ap_q1
).astype(int)

model_with_ASE["risk_highEffort_lowAP"] = (
    (model_with_ASE["high_effort"] == 1) &
    (model_with_ASE["low_AP"] == 1)
).astype(int)

# ============================================================
# 3) High-withdrawal module indicator
# ============================================================

withdraw_by_module = (
    model_with_ASE
    .groupby(["code_module","code_presentation"])["is_withdrawn"]
    .mean()
    .reset_index()
    .rename(columns={"is_withdrawn":"withdraw_rate"})
)

wrate_q3 = withdraw_by_module["withdraw_rate"].quantile(0.75)

high_wrate_modules = withdraw_by_module[
    withdraw_by_module["withdraw_rate"] >= wrate_q3
][["code_module","code_presentation"]]

model_with_ASE["high_withdraw_module"] = model_with_ASE.set_index(
    ["code_module","code_presentation"]
).index.isin(
    high_wrate_modules.set_index(["code_module","code_presentation"]).index
).astype(int)

# ============================================================
# 4) Late Registration + Low VLE usage
# ============================================================

# A) Late registration within module-presentation
reg_q3 = model_with_ASE.groupby(
    ["code_module","code_presentation"]
)["date_registration"].transform(lambda s: s.quantile(0.75))

model_with_ASE["late_registration"] = (
    model_with_ASE["date_registration"] >= reg_q3
).astype(int)

# B) Low VLE usage (totalclicks column from modeldf)
click_col = "total_clicks"  # adjust if your column is named "total_clicks"
click_q1 = model_with_ASE.groupby(
    ["code_module","code_presentation"]
)[click_col].transform(lambda s: s.quantile(0.25))

model_with_ASE["low_vle_usage"] = (
    model_with_ASE[click_col] <= click_q1
).astype(int)

# C) Combined risk
model_with_ASE["risk_late_lowVLE"] = (
    (model_with_ASE["late_registration"] == 1) &
    (model_with_ASE["low_vle_usage"] == 1)
).astype(int)
```

```python
def risk_summary(flag_col):
    return (
        model_with_ASE.groupby(flag_col)["is_withdrawn"]
        .mean()
        .reset_index()
        .rename(columns={"is_withdrawn":"withdraw_rate"})
    )

print("Low Early + Low Submission:\n", risk_summary("risk_lowEarly_lowSub"), "\n")
print("High Effort + Low AP:\n", risk_summary("risk_highEffort_lowAP"), "\n")
print("High-withdraw module:\n", risk_summary("high_withdraw_module"), "\n")
print("Late Registration + Low VLE:\n", risk_summary("risk_late_lowVLE"))
```
